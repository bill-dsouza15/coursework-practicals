Reading data and displaying general information

```python
import pandas as pd
import numpy as np

dataframe = pd.read_csv('DS\\A1\\Dataset\\telecom_churn.csv')
dataframe
```

```python
print("The shape of the dataset is : ")
print(dataframe.shape)
print("\n")
print("The info of each attribute : ")
print(dataframe.info())
```

```python
print("The statistical info of each attribute : ")
dataframe.describe(include = ['int','float','object'])
```

Counting unique values of an attribute

```python
print("The counts of unique records of Churn : ")
dataframe['Churn'].value_counts()
```

Conversion of datatype

```python
print("Converting datatype from bool to int\n")
dataframe['Churn'] = dataframe['Churn'].astype(int)
dataframe.info()
```

Checking null values

```python
print("Check dataframe using notnull()function : ");
print(dataframe.notnull())

print("\nThe count of null values in the dataframe are as follows : ");
print(dataframe.isnull().sum())
```

Indexing, selecting data and sorting data

```python
print("Selection of DAY info in dataset")
day = dataframe.iloc[0:,[0,6,7,8]]
print(day)

print("\nSort day dataframe by Total day charge")
day.sort_values(by=['Total day charge'],ascending=[False],inplace=True)
print("5 highest charges")
print(day.head(5))

print("\n5 lowest charges")
print(day.tail(5))
```

```python
print("Selection of EVE info in dataset")
eve = dataframe.iloc[0:,[0,9,10,11]]
print(eve)

print("\nSort day dataframe by Total eve charge")
eve.sort_values(by=['Total eve charge'],ascending=[False],inplace=True)
print("5 highest charges")
print(eve.head(5))

print("\n5 lowest charges")
print(eve.tail(5))
```

```python
print("Selection of NIGHT info in dataset")
night = dataframe.iloc[0:,[0,12,13,14]]
print(night)

print("\nSort day dataframe by Total night charge")
night.sort_values(by=['Total night charge'],ascending=[False],inplace=True)
print("5 highest charges")
print(night.head(5))

print("\n5 lowest charges")
print(night.tail(5))
```

```python
print("Selection of INTL info in dataset")
intl = dataframe.iloc[0:,[0,15,16,17]]
print(intl)

print("\nSort day dataframe by Total intl charge")
intl.sort_values(by=['Total intl charge'],ascending=[False],inplace=True)
print("5 highest charges")
print(intl.head(5))

print("\n5 lowest charges")
print(intl.tail(5))
```

Identification and treatment of missing values (using another dataset)

```python
dataframe2 = pd.read_csv('DS\\A1\\Dataset\\PimaIndiansDiabetes.csv')
dataframe2
```

```python
dataframe2.info()
```

```python
print("Null count : Before replacing")
print(dataframe2.isnull().sum())

print("\nReplace with mean")
dataframe2.fillna(dataframe2.mean(), inplace=True)
print(dataframe2)

print("\nNull count : After replacing")
print(dataframe2.isnull().sum())
```
