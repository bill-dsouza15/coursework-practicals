Displaying general information

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tset = pd.read_csv('DS\\A5\\Dataset\\titanic.csv')
tset
```

```python
print("Shape of dataset\n",tset.shape)

print("\n\nDataType info\n")
print(tset.info())

print("\nNull values")
print(tset.isnull().sum())
```

Data Visualization

Using Histogram

```python
print("Data visualization using histogram\n")
for x in tset:
  if(np.issubdtype(tset[x].dtype, np.number)):
    tset.hist(column=x, bins=25, grid=False)
```

Using ScatterPlot

```python
print("Data visualization using scatterplot\n")

def plotScatter(xcol, ycol, colour):
  y = tset[ycol]
  x = tset[xcol]
  plt.scatter(x, y, c=colour)
  plt.xlabel(xcol)
  plt.ylabel(ycol)
  plt.title(ycol+" vs "+xcol)
  return plt

plt = plotScatter("Age", "Fare", "blue")
plt.show()
```

Using Pie Chart

```python
tset.Pclass.value_counts(sort=False).plot.pie(autopct='%1.1f%%')
plt.title("Pclass Distribution")
plt.show()

tset.Sex.value_counts(sort=False).plot.pie(autopct='%1.1f%%')
plt.title("Sex")
plt.show()

tset.Embarked.value_counts(sort=False).plot.pie(autopct='%1.1f%%')
plt.title("Embarked")
plt.show()

tset.Survived.value_counts(sort=False).plot.pie(autopct='%1.1f%%')
plt.title("Survived")
plt.show()
```

Using BarPlot

```python
print("Data visualization using Barplot\n")

sns.set()
def bar_chart(feature):
    survived = tset[tset['Survived']==1][feature].value_counts()
    not_survived = tset[tset['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,not_survived])
    df.index=['Survived','Not_survived']
    df.plot(kind='bar',stacked=True,title=feature)

bar_chart("Sex")
bar_chart("Pclass")
bar_chart("Embarked")
bar_chart("SibSp")
bar_chart("Parch")
```

Using BoxPlot

```python
print("Data visualization using Barplot\n")

sns.set()
def printBoxPlot(xd, yd):
  sns.boxplot(x=xd, y=yd, data=tset)
  plt.show()

printBoxPlot("Survived", "Age")
printBoxPlot("Survived", "Fare")
```
